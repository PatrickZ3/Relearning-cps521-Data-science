```python
#Exercise 3
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = pd.read_csv("Dataset student_score.csv")

x = data.Xi
y = data.Yi

x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)

reg = LinearRegression().fit(x, y)

r_squared = reg.score(x, y)

y_pred = reg.predict(x)

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

baseline = np.var(y)

print("R-squared value:", r_squared)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Baseline:", baseline)

```

```python
# Exercise 5
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score

data = pd.read_csv('heart-disease-dataset.csv')

for col in data.columns:
    data = data[data[col] != '?']
    
correlation_matrix = data.corr().round(2)
sbn.set(rc={'figure.figsize': (11.7, 8.27)})
sbn.heatmap(data=correlation_matrix, annot=True) 
```

```python
x_oldpeak = data['oldpeak']
y_result = data['result']

x_oldpeak = np.array(x_oldpeak).reshape(-1,1)
y_result = np.array(y_result).reshape(-1,1)

x_train1, x_test1, y_train1, y_test1 = train_test_split(x_oldpeak, y_result, test_size=0.2,random_state=3)

lr = linear_model.LinearRegression()
lr.fit(x_train1, y_train1)

y_pred1 = lr.predict(x_test1)

mse = mean_squared_error(y_test1, y_pred1)
r_squared = round(lr.score(x_test1, y_test1), 2)

print('Model Performance')
print('___________________________')
print(f'Mean Squared Error: {mse}')
print(f'R-Squared Value: {r_squared}')

pred_space = np.linspace(min(x_oldpeak), max(x_oldpeak)).reshape(-1,1)
plt.scatter(x_oldpeak, y_result)
plt.plot(pred_space, lr.predict(pred_space), color = 'red', linewidth=4)
plt.xlabel('old peak')
plt.ylabel('result')
plt.show()
```
