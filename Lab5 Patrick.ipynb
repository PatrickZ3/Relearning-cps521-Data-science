```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv('heart-disease-dataset.csv')
file = file.replace({'?' : np.nan})
file.dropna(inplace=True)

k_means = KMeans(n_clusters= 5)

y_k_means = k_means.fit_predict(file[['age', 'chol']])
file['cluster'] = y_k_means

plt.scatter(file[file.cluster==0].age, file[file.cluster==0].chol,  c='red', label='0')
plt.scatter(file[file.cluster==1].age, file[file.cluster==1].chol,  c='blue', label='1')
plt.scatter(file[file.cluster==2].age, file[file.cluster==2].chol,  c='yellow', label='2')
plt.scatter(file[file.cluster==3].age, file[file.cluster==3].chol,  c='purple', label='3')
plt.scatter(file[file.cluster==4].age, file[file.cluster==4].chol,  c='green', label='4')
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv('heart-disease-dataset.csv')
file = file.replace({'?' : np.nan})
file.dropna(inplace=True)

k_means = KMeans(n_clusters= 3)

y_k_means = k_means.fit_predict(file[['age', 'chol']])
file['cluster'] = y_k_means

plt.scatter(file[file.cluster==0].age, file[file.cluster==0].chol,  c='red', label='0')
plt.scatter(file[file.cluster==1].age, file[file.cluster==1].chol,  c='blue', label='1')
plt.scatter(file[file.cluster==2].age, file[file.cluster==2].chol,  c='yellow', label='2')
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

file = pd.read_csv('heart-disease-dataset.csv')
file = file.replace({'?' : np.nan})
file.dropna(inplace=True)

plt.figure(figsize=(10,6))
dend = sch.dendrogram(sch.linkage(file[['age','chol']],method = 'ward'))

plt.title('Dendogram')
plt.xlabel('Sample')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

file = pd.read_csv('heart-disease-dataset.csv')
file = file.replace({'?' : np.nan})
file.dropna(inplace=True)


ac = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_ac = ac.fit_predict(file[['age', 'chol']])

file['cluster'] = y_ac


plt.scatter(file[file.cluster==0].age, file[file.cluster==0].chol,  c='red', label='0')
plt.scatter(file[file.cluster==1].age, file[file.cluster==1].chol,  c='blue', label='1')
plt.scatter(file[file.cluster==2].age, file[file.cluster==2].chol,  c='yellow', label='2')
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```
